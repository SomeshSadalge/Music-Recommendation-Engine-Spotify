# Building Recommendation Engine

## Step 1 - Find the Euclidian Distance between given user and all the other user
## Step 2 - Find 5 closest user
## Step 3 - Find songs that these 5 closest user listen to
## Step 4 - Recommend these songs to user

```python
def euclidian_dist(v1,v2):
    return pow(sum([pow(v1[idx]-v2[idx],2) for idx in range(len(v1))]),0.5)
```

```python
euclidian_dist(user_matrix[0],user_matrix[1])
```

```python
user1_vec = user_matrix[0]
```

```python
eucl_dist = []
```

```python
for idx in range(1000):
    other_user = user_matrix[idx]
    eu = euclidian_dist(user1_vec,other_user)
    eucl_dist.append(eu)
```

```python
candidate_index = np.argsort(eucl_dist)
```

```python
candidate_index[1:6]
```

# Finding Eucledian distance between the given users to all the users

```python
# Finding Euclidian distance between given user to all user
import sys
eucl_dist1 = list()
user1_id = 0 
counter = 0
```

```python
for any_user in user_matrix[:]:
    if user1_id == counter:
        eucl_dist1.append(sys.float_info.max)
    else:
        user1 = user_matrix[0]
        dist = euclidian_dist(user1, any_user)
        eucl_dist1.append(dist)
counter += 1

eucl_dist1
```

```python
closest_user_indices = [np.argsort(eucl_dist1)[:5]]
closest_user_indices
```

```python
# What songs do this 5 people listen to
for idx in closest_user_indices:
    temp = pd.DataFrame(df.iloc[idx])
    print(temp[:].index)
temp.head()
```

# Implimentaion of K-means

```python
# Assume that user1 is listening to song1
# Following are steps for K-means
# Define the number of cluster for different song k=10
```

```python
k= 10
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= k)
kmeans.fit(song_matrix)
```

```python
# Find which cluster song belong
# Find vector for song 1
indexin_df = list(df.columns).index('song_1')
```

```python
print(indexin_df)
```

```python
song1_vec = song_matrix[indexin_df]
kmeans.predict([song1_vec])
```

```python
song_matrix.shape
```

```python
cluster_id = 8
allLabels = list(kmeans.predict(song_matrix))
predicted_cluster_id = [idx for idx in range(len(allLabels)) if allLabels[idx == cluster_id]]
predicted_cluster_candidate = song_matrix[predicted_cluster_id]
print(predicted_cluster_candidate)
```

```python
from sklearn.neighbors import NearestNeighbors
KNN = NearestNeighbors(n_neighbors= 5)
KNN.fit(predicted_cluster_candidate)
```

```python
KNN.kneighbors([song_matrix[indexin_df]])
```
